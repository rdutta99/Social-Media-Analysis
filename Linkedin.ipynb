```python
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv("USEReady1.csv", index_col=False)
df.head()
```

```python
df['timestamp']=df['timestamp'].str.replace("T"," ")
df['timestamp']=df['timestamp'].str.replace("Z"," ")
df.head()
```

```python
df['timestamp']=pd.to_datetime(df['timestamp'])
type(df['timestamp'][0])
```

```python
df.head(30)
```

```python
print(df['timestamp'][0]-pd.to_datetime(15))
```

```python
def date():
    l=[]
    for i in range(len(df)):
        print(i)
        x=df['timestamp'][i]
        s=df['postDate'][i]
        if s[1]=='h':
            l.append(x - pd.DateOffset(hours=int(s[0])))
        elif s[2]=='h':
            l.append(x - pd.DateOffset(hours=int(s[0])*10+int(s[1])))
        elif s[1]=='d':
            l.append(x - pd.DateOffset(days=int(s[0])))
        elif s[2]=='d':
            l.append(x - pd.DateOffset(days=int(s[0])*10+int(s[1])))
        elif s[1]=='w':
            l.append(x - pd.DateOffset(weeks=int(s[0])))
        elif s[1]=='m':
            l.append(x - pd.DateOffset(months=int(s[0])))
        elif s[2]=='m':
            l.append(x - pd.DateOffset(months=int(s[0])*10+int(s[1])))
        else:
            l.append(0)
    return l
date=date()
df['date']=date
df.head()
```

```python
df = df[df['date'] != 0]
len(df)
```

```python
df['date']=[d.date() for d in df['date']]
del df['timestamp']
df.head()
```

```python
del df['postDate']
del df['query']
del df['postUrl']
```

```python
df.to_csv("USEReady_Linkedin.csv",index=False)
```
